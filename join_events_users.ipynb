```python
import pandas as pd

users = pd.read_csv('/content/users.csv')
events = pd.read_csv('/content/events.csv')
```

```python
users.head()
```

```python
events.head()
```

```python
events['created_at'] = pd.to_datetime(events['created_at'], errors='coerce')
users['created_at'] = pd.to_datetime(users['created_at'], errors='coerce')
```

```python
events.dtypes
```

```python
events_users = pd.merge(events, users,
                             left_on=['user_id'],
                             right_on=['id'],
                             how='left',
                             suffixes=('_even', '_users'))
```

```python
events_users.dtypes
```

```python
events_users = events_users.drop(columns=['traffic_source_users', "postal_code_users", "city_users", "state_users", "created_at_users", "id_users", "postal_code_users"])
```

```python
events_users.rename(columns={'created_at_even': 'created_at', 'city_even': 'city', 'state_even': 'state', 'postal_code_even': 'postal_code', 'traffic_source_even': 'traffic_source', 'state_users': 'state'}, inplace=True)
```
